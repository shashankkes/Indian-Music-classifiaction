```python

def features_extractor(file):
  y, sr = librosa.load(file)
  S = np.abs(librosa.stft(y))
  data=[]
  tempo, beats = librosa.beat.beat_track(y=y, sr=sr)
  chroma_stft = librosa.feature.chroma_stft(y=y, sr=sr)
  chroma_cq = librosa.feature.chroma_cqt(y=y, sr=sr)
  chroma_cens = librosa.feature.chroma_cens(y=y, sr=sr)
  melspectrogram = librosa.feature.melspectrogram(y=y, sr=sr)
  cent = librosa.feature.spectral_centroid(y=y, sr=sr)
  spec_bw = librosa.feature.spectral_bandwidth(y=y, sr=sr)
  contrast = librosa.feature.spectral_contrast(S=S, sr=sr)
  rolloff = librosa.feature.spectral_rolloff(y=y, sr=sr)
  poly_features = librosa.feature.poly_features(S=S, sr=sr)
  tonnetz = librosa.feature.tonnetz(y=y, sr=sr)
  zcr = librosa.feature.zero_crossing_rate(y)
  harmonic = librosa.effects.harmonic(y)
  percussive = librosa.effects.percussive(y)

  mfcc = librosa.feature.mfcc(y=y, sr=sr)
  mfcc_delta = librosa.feature.delta(mfcc)

  onset_frames = librosa.onset.onset_detect(y=y, sr=sr)
  frames_to_time = librosa.frames_to_time(onset_frames[:20], sr=sr)
  data.append(tempo)
  data.append(sum(beats))
  data.append(np.average(beats))
  data.append(np.mean(chroma_stft))
  data.append(np.std(chroma_stft))
  data.append(np.var(chroma_stft)) 
  data.append(np.mean(chroma_cq))
  data.append(np.std(chroma_cq))
  data.append(np.var(chroma_cq))
  data.append(np.mean(chroma_cens))
  data.append(np.std(chroma_cens))
  data.append(np.var(chroma_cens))
  data.append(np.mean(melspectrogram))
  data.append(np.std(melspectrogram))
  data.append(np.var(melspectrogram))
  data.append(np.mean(mfcc))
  data.append(np.std(mfcc))
  data.append(np.var(mfcc))
  data.append(np.mean(mfcc_delta))
  data.append(np.std(mfcc_delta))
  data.append(np.var(mfcc_delta))
  data.append(np.mean(cent))
  data.append(np.std(cent))
  data.append(np.var(cent))
  data.append(np.mean(spec_bw))
  data.append(np.std(spec_bw))
  data.append(np.var(spec_bw))
  data.append(np.mean(contrast))
  data.append(np.std(contrast))
  data.append(np.var(contrast))
  data.append(np.mean(rolloff))
  data.append(np.std(rolloff))
  data.append(np.var(rolloff))
  data.append(np.mean(poly_features))
  data.append(np.std(poly_features))
  data.append(np.var(poly_features))
  data.append(np.mean(tonnetz))
  data.append(np.std(tonnetz))
  data.append(np.var(tonnetz))
  data.append(np.mean(zcr))
  data.append(np.std(zcr))
  data.append(np.var(zcr))
  data.append(np.mean(harmonic))
  data.append(np.std(harmonic))
  data.append(np.var(harmonic))
  data.append(np.mean(percussive))
  data.append(np.std(percussive))
  data.append(np.var(percussive))
  data.append(np.mean(frames_to_time))
  data.append(np.std(frames_to_time))
  data.append(np.var(frames_to_time))
  return data
```

```python
import numpy as np
from tqdm import tqdm
### Now we iterate through every audio file and extract features 
### using Mel-Frequency Cepstral Coefficients
extracted_features=[]
for index_num,row in tqdm(metadata.iterrows()):
    file_name = "/content/drive/MyDrive/Song_Dataset/{}.wav".format(index_num)
    final_class_labels=row["Class"]
    data=features_extractor(file_name)
    extracted_features.append([data,final_class_labels])
```

```python
### converting extracted_features to Pandas dataframe
extracted_features_df=pd.DataFrame(extracted_features,columns=['feature','class'])
extracted_features_df.head()
```

```python
extracted_features_df.shape
```

```python
X=np.array(extracted_features_df['feature'].tolist())
y=np.array(extracted_features_df['class'].tolist())
```

```python
from tensorflow.keras.utils import to_categorical
from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
y=to_categorical(labelencoder.fit_transform(y))
```

```python
### Train Test Split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout,Activation,Flatten
from tensorflow.keras.optimizers import Adam
from sklearn import metrics
```

```python
num_labels=y.shape[1]
```

```python
model=Sequential()
###first layer
model.add(Dense(256,input_shape=(51,)))
model.add(Activation('relu'))
model.add(Dropout(0.5))
###second layer
model.add(Dense(512))
model.add(Activation('relu'))
model.add(Dropout(0.5))
###third layer
model.add(Dense(512))
model.add(Activation('relu'))
model.add(Dropout(0.5))
###forth layer
model.add(Dense(512))
model.add(Activation('relu'))
model.add(Dropout(0.5))
###fifth layer
model.add(Dense(512))
model.add(Activation('relu'))
model.add(Dropout(0.5))
###sixth layer
model.add(Dense(512))
model.add(Activation('relu'))
model.add(Dropout(0.5))
###seventh layer
model.add(Dense(512))
model.add(Activation('relu'))
model.add(Dropout(0.5))
###eighth layer
model.add(Dense(512))
model.add(Activation('relu'))
model.add(Dropout(0.5))
###final layer
model.add(Dense(num_labels))
model.add(Activation('softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='categorical_crossentropy',metrics=['accuracy'],optimizer='adam')
```
